```python
import numpy as np
import scipy.linalg
import matplotlib.pyplot as plt
```

```python
# create 101 evenly spaced points between -10 and 10
x = np.linspace(-10,10,101)
x
```

```python
# create a function of our grid points
y = np.cos(3*x)*np.exp(-0.1*abs(x))
```

```python
# plot our data
plt.plot(x,y);
```
