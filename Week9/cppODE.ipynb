```c++
#include <iostream>
#include <cmath>
#include "Eigen/Dense"
```

```c++
using namespace Eigen;
```

```c++
int N = 100;
VectorXd x(N), y(N);
MatrixXd d2ydx2(N, N);
```

```c++
double dx = 1.0/(99.0);
for (int i = 0; i < 100; i++) {
    x(i) = 0 + i*dx;
}
```

```c++
std::cout << x;
```

```c++
VectorXd g = x*dx*dx
```

```c++
g(0) = -2*0.2;
g(99) = -2*1.5 ;
std::cout << g;
```

```c++
for (int i = 0; i < 100; i++) {
    for (int j = 0; j < 100; j++) {
        d2ydx2(i,j) = 0.0;
    }
}

for (int i = 1; i < 99; i++) {
    d2ydx2(i,i) = -2;
    d2ydx2(i,i-1) = 1;
    d2ydx2(i,i+1) = 1;
}
d2ydx2(0,0) = -2;
d2ydx2(99,99) = -2;
std::cout << d2ydx2;
```

```c++
y = d2ydx2.fullPivLu().solve(g);
```

```c++
std::cout << y;
```
