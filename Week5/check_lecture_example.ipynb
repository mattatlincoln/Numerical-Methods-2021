```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([-3.0,-2.3,-1.6,-0.9,-0.2,0.5,1.2,1.9,2.6,3.3,4.0])
y = np.array([8.26383742,6.44045188,4.74903073,4.5656476,3.61011683,
              3.32743918,2.9643915,1.02239181,1.09485138,1.84053372,1.49110572])
plt.scatter(x,y, marker = 'x')
plt.xlabel(r'$x_i$')
plt.ylabel(r'$y_i$');
```

```python
Z = np.zeros((len(x),3))
# this uses numpy's vectorization and calculates them for every value of x
# it is the same as having a loop over the x values
Z[:,0] = 1
Z[:,1] = np.exp(-x)
Z[:,2] = np.exp(-2*x)
print(Z, "\n")
print(Z.transpose())
```

```python
a = np.linalg.solve(np.dot(Z.transpose(), Z), np.dot(Z.transpose(),y))
a
```

```python
def model(xi, a):
    return a[0] + a[1]*np.exp(-xi) + a[2]*np.exp(-2*xi)

plt.scatter(x,y, marker = 'x')
plt.plot(x, model(x,a))
plt.xlabel(r'$x_i$')
plt.ylabel(r'$y_i$');
```
